```python
titanic_data=titanic_data.drop(columns=['Name','Ticket'],axis=1)
titanic_data.head()
```

```python
titanic_data["Sex"].value_counts()
```

```python
titanic_data["Embarked"].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
cols =['Sex','Embarked']
le =LabelEncoder()
for col in cols:
    titanic_data[col] =le.fit_transform(titanic_data[col])
titanic_data.head()
```

Train_Test_Split

```python
X=titanic_data.drop(columns =['PassengerId','Survived'],axis= 1)
Y =titanic_data['Survived']
```

```python
print(X)
```

```python
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.2,random_state =2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model=LogisticRegression()
```

```python
titanic_data.info()
```

```python
titanic_data.astype({'Age':'int','Fare':'int'}).dtypes
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction =model.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy =accuracy_score(Y_train,X_train_prediction)
print("Accuracy score of training model:",training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuarcy =accuracy_score(Y_test,X_test_prediction)
print('Accuarcy score of test data :',test_data_accuarcy)
```
